# txt 수정하기

```python
import os
import pandas as pd
```

```python
path = '../data__/train_annotation.txt'

df = pd.read_csv(path, header=None)
df.head()
```

```python
new_f = open('../data__/train_annotation__.txt', mode='w')

with open('../data__/train_annotation.txt', mode='r') as f:
    data = f.readlines()
    
    for i in range(len(data)):
        data_lst = data[i].split(',')
        title = data_lst[0]
        title_ = title.replace('data__', 'data')        
        val = title_.split(' ')
        data_lst = val + data_lst[1:]
        res = ''
        for j in range(len(data_lst)):
            res += data_lst[j]
            if j % 5 == 0:
                res += ' '
            else:
                res += ','
                    
        new_f.write(res[:-1])
    
new_f.close()
```

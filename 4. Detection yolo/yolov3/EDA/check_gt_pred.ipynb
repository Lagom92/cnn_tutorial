# Check GT and predict

```python
# 라이브러리
import os
import cv2
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"]=10, 10
```

```python
test_path = '../data/test/'
test_data_lst = os.listdir(test_path)

print(len(test_data_lst))

txt_path = '../data/test_annotation.txt'

df = pd.read_csv(txt_path, header=None)
print(df.shape)
```

```python
df
```

```python
n = len(test_data_lst)

for i in range(n):
    # image
    img_path = test_path + test_data_lst[i]
    img = cv2.imread(img_path)

    sr = df.iloc[i].tolist()
    # bounding box
    image_path, x = sr[0].split(' ')
    boxes = [x]
    for i in range(1, len(sr)):
        try:
            boxes.extend(sr[i].split(' '))
        except:
            boxes.append(sr[i])
    for i in range(0, 30, 5):
        x1, y1, x2, y2, c = int(boxes[i]), boxes[i+1], boxes[i+2], boxes[i+3], boxes[i+4]

        img = cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 1)
        cv2.putText(img, str(c), (x1, y1), cv2.FONT_ITALIC, 0.7, (0, 0, 255), 1, cv2.LINE_AA)
        
    cv2.imwrite('../data/gt/'+image_path.split('/')[-1], img)
#     plt.imshow(img)
#     plt.show()
print("Success")
```
